# Project 3: House Price Explorer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)
df=pd.read_csv(r"C:/Users/nithy/OneDrive/Desktop/Training/Level_5/train.csv")
df.head()
```

```python
df["TotalSqFeet"] = df[["TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "LowQualFinSF"]].fillna(0).sum(axis=1)
df.head()
```

```python
summary=df[['SalePrice','TotalSqFeet']].describe()
print("summary statistics for price and square footage")
summary
```

```python
df["price_per_sqft"]=df["SalePrice"]/df["TotalSqFeet"]
df.head()
```

```python
plt.scatter(x=df['TotalSqFeet'],y=df['price_per_sqft'])
plt.title("Price per square feet")
plt.xlabel("Sqaure feet")
plt.ylabel("Price per feet")
plt.grid(True)
plt.show()
```

```python
plt.scatter(x=df['TotalSqFeet'],y=df['SalePrice'])
plt.title("Price per square feet")
plt.xlabel("Sqaure feet")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```
