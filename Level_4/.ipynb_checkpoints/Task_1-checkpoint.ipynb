## Task 1: House Price EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
df=pd.read_csv(r"C:/Users/nithy/OneDrive/Desktop/Training/Level_4/train.csv")
df.head(10)
```

```python
df.shape
```

```python
null=df.isnull().sum()
null
```

```python
features_na=[features for features in df.columns if df[features].isnull().sum()>1]
for feature in features_na:
    print(feature,np.round(df[feature].isnull().mean(),4))
```

```python
for feature in features_na:
    df1=df.copy()
    df1[feature]=np.where(df1[feature].isnull(),1,0)
    df2=df1.groupby(feature)['SalePrice'].median()
    df2=df2.to_frame()
    df2=df2.reset_index()
    sns.barplot(x=df2[feature], y=df2["SalePrice"], hue=df2[feature])
    plt.title(feature)
    plt.show()
```

```python
numerical_feature=[feature for feature in df.columns if df[feature].dtypes!='O']
print("Numerical_Feature:",len(numerical_feature))
df[numerical_feature].head()
```

```python
year_feature=[feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
df3=df.groupby('YrSold')['SalePrice'].median()
df3=df3.to_frame()
df3=df3.reset_index()
df3
```

```python
sns.lineplot(x=df3['YrSold'], y=df3['SalePrice'],marker='o')
plt.title("House Price Vs YearSold")
plt.xlabel("Year")
plt.ylabel("House Price")
plt.show()
```

```python
for feature in year_feature:
    if feature!='YrSold':
        df4=df.copy()
        df4[feature]=df4['YrSold']-df4[feature]
        sns.scatterplot(x=df4[feature], y=df4['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
df4.head()
```

```python
discrete_feature=[feature for feature in numerical_feature if len(df[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Feature Count={}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
for feature in discrete_feature:
    df5=df.copy()
    df5=df5.groupby(feature)['SalePrice'].median() 
    df5=df5.to_frame()
    df5=df5.reset_index()
    sns.barplot(x=feature,y='SalePrice',data=df5,hue=feature)
    plt.title(feature)
    plt.show()
```

```python
continuous_feature=[feature for feature in numerical_feature if feature not in discrete_feature+['Id']]
print("Continuous Feature Count={}".format(len(continuous_feature)))
```

```python
continuous_feature
```

```python
for feature in continuous_feature:
    df6=df.copy()
    df6[feature].hist(bins=25)
    plt.title(feature)
    plt.xlabel(feature)
    plt.ylabel('counts')
    plt.show()
```

### logarithmic transformation

```python
for feature in continuous_feature:
    df6=df.copy()
    if 0 in df6[feature].unique():
        pass
    else:
        df6[feature]=np.log(df6[feature])
        df6['SalePrice']=np.log(df6['SalePrice'])
        plt.scatter(df6[feature],df6['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

### Outliers


```python
for feature in continuous_feature:
    df7=df.copy()
    if 0 in df7[feature].unique():
        pass
    else:
        df7[feature]=np.log(df7[feature])
        df7['SalePrice']=np.log(df7['SalePrice'])
        df7.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

### Categorical Variables

```python
categorical_features=[feature for feature in df.columns if df[feature].dtypes=='O']
categorical_features
```

```python
for feature in categorical_features:
    print("Feature is {} and the number of categories are {}".format(feature,df[feature].nunique()))
```

```python
for feature in categorical_features:
    df8=df.copy()
    df8=df8.groupby(feature)['SalePrice'].median()
    df8=df8.to_frame()
    df8=df8.reset_index()
    sns.barplot(x=feature,y='SalePrice',data=df8,hue=feature)
    plt.title(feature)
    plt.show()
```
