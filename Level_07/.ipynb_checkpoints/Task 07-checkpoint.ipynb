## Task 7: Simulate Bernoulli Trials

```python
import numpy as np
```

```python
np.random.seed(123)
bi=np.random.binomial(n=1,p=0.3,size=1000)
bi
```

```python
p=0.3
exp_p=np.round(bi.mean(),1)
print(f"Expected success probability : {exp_p}")
```

```python
if exp_p==p:
    print("Matches theoretical P = 0.3 for Bernoulli trial.")
else:
    print("Does not match theoretical P = 0.3 for Bernoulli trial.")
```
