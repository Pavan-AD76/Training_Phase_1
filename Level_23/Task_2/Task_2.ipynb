```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel("C:/Users/nithy/OneDrive/Desktop/Training/Level_23/Task_2/DATA.xlsx")  
```

```python
income_df = df[['YEAR', 'Average', 'MIN', 'MAX']]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(income_df['YEAR'], income_df['Average'], label='Average Income', color='blue')
plt.plot(income_df['YEAR'], income_df['MIN'], label='Min Income', color='green')
plt.plot(income_df['YEAR'], income_df['MAX'], label='Max Income', color='red')
plt.title('Income Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Income (in international dollars per day)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
