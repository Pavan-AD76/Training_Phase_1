```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("C:/Users/nithy/OneDrive/Desktop/Training/Level_23/Task_1/Data.xlsx")
```

```python
df_life_exp = df.set_index('country').transpose()
df_life_exp.index = df_life_exp.index.astype(int)
avg_life_exp = df_life_exp.mean(axis=1, skipna=True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(avg_life_exp.index, avg_life_exp.values, color='blue', linewidth=2)
plt.title('Average Life Expectancy Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Life Expectancy')
plt.grid(True)
plt.tight_layout()
plt.show()
```
