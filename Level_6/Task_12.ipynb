# Task 12: QQ Plot

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data1=np.random.normal(0,1,1000)
```

```python
data1=np.asarray(data1)
```

```python
stats.probplot(data1,dist="norm",plot=plt)
plt.show()
```
